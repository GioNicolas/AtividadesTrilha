## Questão 6 📝

Crie um programa que recebe o salário de um funcionário e calcula o imposto devido com base nas faixas abaixo:

| Faixa salarial | Alíquota |
|---|---|
| Até R\$ 1.500 | Isento (0%) |
| R\$ 1.500,01 – R\$ 3.000 | 7,5% |
| R\$ 3.000,01 – R\$ 4.500 | 15% |
| Acima de R\$ 4.500 | 22,5% |

Exiba o salário bruto, o valor do imposto e o salário líquido.

````python
salario = float(input("Digite seu salário: "))

if salario <= 1500:
  imposto = 0
elif salario <= 3000:
  imposto = salario * (7.5/100)
elif salario <= 4500:
  imposto = salario * (15/100)
else:
  imposto = salario * (22.5/100)


salario_liquido = salario - imposto
print(f"Seu salário bruto é {salario} que possui {imposto} a ser pago, resultado em {salario_liquido} de salário líquido")
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Use `if`/`elif`/`else` para cada faixa. Calcule `imposto = salario * aliquota`, depois `liquido = salario - imposto`. *

## Questão 7 📝

Crie um programa que verifica se um ano digitado pelo usuário é bissexto.

**Regras:**
- Divisível por 4 → bissexto
- Mas se divisível por 100 → **NÃO** é bissexto
- A menos que também seja divisível por 400 → **É** bissexto

Exiba `"Ano bissexto"` ou `"Não é bissexto"`.

````python
ano = int(input("Digite o ano: "))

if ano %  4 == 0 and ano % 100 != 0 or ano % 400 == 0:
  print("O ano é bissexto")
else:
  print("O ano não é bissexto")
````

## Questão 8 📝

Construa um programa que recebe o peso e a altura do usuário, calcula o IMC e classifica o resultado:

| IMC | Classificação |
|---|---|
| IMC < 18,5 | Abaixo do peso |
| 18,5 <= IMC < 25 | Peso normal |
| 25 <= IMC < 30 | Sobrepeso |
| IMC >= 30 | Obesidade |

Exiba o IMC com 2 casas decimais e a classificação.

````python
peso = float(input("Digite seu peso: "))
altura = float(input("Digite sua altura em metros: "))

imc =  peso / (altura **2)

if imc < 18.5:
  classe = ("Abaixo do peso")
elif imc <= 25:
  classe = ("Peso normal")
elif imc < 30:
  classe = ("Sobrepeso")
else:
  classe = ("Obesidade")

print(f"Seu IMC é {imc:.2f} e sua classificação é = {classe} ")

````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*`IMC = peso / altura**2`. Para formatar com 2 casas decimais, use `f"{imc:.2f}"`.*

## Questão 9 📝

Peça ao usuário para digitar uma palavra e use métodos de string para:
1. Exibir com a primeira letra maiúscula
2. Exibir quantas letras ela tem
3. Exibir a palavra de trás para frente (slicing reverso)
4. Verificar se ela começa com a letra `"a"` (sem diferenciar maiúsculas de minúsculas)

````python
palavra = input("Digite a palavra: ")

print(palavra.capitalize())

letras = len(palavra)
print(f"Possui {letras} letras.")

print(palavra[::-1])

if palavra.lower().startswith("a"):
  print("Sua palavra começa com a.")
else:
  print("Sua palavra NAO começa com a.")

````

### 💡 Dica:

## Questão 10 📝

Crie um classificador de triângulos. Receba os 3 lados pelo `input()` e:

1. Verifique se os lados formam um triângulo válido *(a soma de dois lados quaisquer deve ser maior que o terceiro)*
2. Se válido, classifique:
   - **Equilátero:** todos os lados iguais
   - **Isósceles:** dois lados iguais
   - **Escaleno:** todos diferentes

Exiba a classificação ou `"Não forma um triângulo"`.

````python
lado1 = input("Digite a medida do primeiro lado: ")
lado2 = input("Digite a medida do segundo lado: ")
lado3 = input("Digite a medida do terceiro lado: ")

if lado1 + lado2 > lado1 and lado2 + lado3 > lado1 and lado1 + lado3 > lado2:
  if lado1 == lado2 == lado3:
    print("Seu triangulo é equilátero")
  elif lado1 == lado2 or lado2 == lado3 or lado1 == lado3:
    print("Seu triangulo é isósceles")
  else:
    print("Seu traingulo é escaleno")
else:
  print("Não forma um triângulo")



````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Condição de triângulo válido: `a+b>c AND a+c>b AND b+c>a`. Depois compare os lados com `==` para classificar.*
